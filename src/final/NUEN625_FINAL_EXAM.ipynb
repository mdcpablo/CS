A 1D spherical $S_N$ code is silimilar to 1D cartesian $S_N$ code, except that one must deal with the angular derivative term $\partial \psi/ \partial \mu$. The idea behind spherical $S_N$ is the same as cartesian $S_N$, you want to solve the transport equation at a lot of different directions, $\mu_m$, and these directions are generated by a quadrature set (such as Gauss-Legendre quadrature). The weights $w_m$ in the quadrature set can be used to determine the values of $\mu_m$. You start my setting $\mu_{1/2} = -1$ and then setting each $\mu_{m+1/2} = \mu_{m-1/2} + w_m$. The values for $\mu_{m-1/2}$ and $\mu_{m+1/2}$ can then be used for a discrtete approximation of the angular derivative term $\partial \psi_m/ \partial \mu$. 

Spatial discrtezation in spherical geometries is different than in cartesian geometries because the volume-averaged flux over a cell, $\psi_i$, is not equal to the arithmetic average of the flux on the two faces, $\psi_{i-1/2}$ and $\psi_{i+1/2}$. In addition, the quadrature points (directions) $\mu_m$ are not necessarily equal to the midpoints of $\mu_{m-1/2}$ and $\mu_{m+1/2}$. The 1D spherical $S_N$ equation with weighted-diamond differencing is
\begin{equation}
\mu_m(A_{i+1/2} \psi_{i+1/2,m} - A_{i-1/2} \psi_{i-1/2,m} ) + \frac{1}{2}(A_{i+1/2} - A_{i-1/2}) \frac{(\alpha_{m+1/2} \hat{\psi}_{i,m+1/2} - \alpha_{m-1/2} \hat{\psi}_{i,m-1/2})}{w_m} + \sigma_t V_i \bar{\psi}_{i,m} = Q_{i,m} V_i
\end{equation}
where
\begin{equation}
\hat{\psi}_{i,m} = \hat{\gamma_i} \psi_{i+1/2,m} + (1-\hat{\gamma_i}) \psi_{i-1/2,m}
\end{equation}
\begin{equation}
\bar{\psi}_{i,m} = \gamma_i \psi_{i+1/2,m} + (1-\gamma_i) \psi_{i-1/2,m}
\end{equation}
\begin{equation}
\hat{\psi}_{i,m} = \beta_m \hat{\psi}_{i,m+1/2} + (1-\beta_m) \hat{\psi}_{i,m-1/2} 
\end{equation}
where $\hat{\gamma_i}$, $\gamma_i$, and $\beta_m$ are used to properly weight the angular flux by $r$, $r^2$, and $\mu$, respectively.

In order to solve the 1D spherical $S_N$ equation, we must sweep in angle and space. An important characteristic of neutrons streaming in spherical geometry is that the direction $\mu$ with respect to the origin is always increasing as the neutron travels. Thus, when sweeping through the spherical geometry we start with small angle $\mu=-1$ and sweep in space for that angle. Then we move onto the next smallest angle and sweep in space for that angle, and so on.  

By sweeping from the smaller angle to larger angle, we know the value for $\hat{\psi}_{i,m-1/2}$ because it comes from the previous-angle's sweep. In addition, when we're sweeping spatially towards $r=0$ we know the value of $\psi_{i+1/2,m}$, and when we're sweeping spatially towards $r=R$ we know the value of $\psi_{i-1/2,m}$. Thus for any given solve there are two known values, which leaves us with four unknowns and four equations. 

To solve for $\hat{\gamma_i}$, we need to make both sides of the following equivalent
\begin{equation}
\hat{\gamma_i} \psi_{i+1/2} + (1-\hat{\gamma_i}) \psi_{i-1/2} = \frac{\int_{r_{i-1/2}}^{r_{i+1/2}} dr \, r \tilde{\psi}}{\int_{r_{i-1/2}}^{r_{i+1/2}} dr \, r}
\end{equation}
where 
\begin{equation}
\tilde{\psi} = \psi_{i+1/2} \Big( \frac{r-r_{i-1/2}}{h_i} \Big) + \psi_{i-1/2} \Big( \frac{r_{i+1/2}-r}{h_i} \Big)  \: .
\end{equation}

This yields that
\begin{equation}
\hat{\gamma_i} \psi_{i+1/2} + (1-\hat{\gamma_i}) \psi_{i-1/2}  = \frac{1}{\Big(\frac{r_{i+1/2}^2-r_{i-1/2}^2}{2}\Big)} \int_{r_{i-1/2}}^{r_{i+1/2}} dr \, \Bigg[ \psi_{i+1/2} \Big( \frac{r^2-r_{i-1/2}r}{h_i} \Big) + \Big( \psi_{i-1/2} \frac{r_{i+1/2}r-r^2}{h_i} \Big) \Bigg]
\end{equation}
\begin{equation}
\hat{\gamma_i} \psi_{i+1/2} + (1-\hat{\gamma_i}) \psi_{i-1/2}  = \frac{2}{h_i \Big(r_{i+1/2}^2-r_{i-1/2}^2 \Big)}  \Bigg[ \Big(\psi_{i+1/2} - \psi_{i-1/2} \Big) \frac{r_{i+1/2}^3 - r_{i-1/2}^3}{3} \, + \, \Big(\psi_{i-1/2}r_{i+1/2} - \psi_{i+1/2}r_{i-1/2} \Big) \frac{r_{i+1/2}^2 - r_{i-1/2}^2}{2} \Bigg] \: .
\end{equation}

Now, by combining the terms that have a $\psi_{i+1/2}$, we can solve for $\gamma_{i,1}$ and get that 
\begin{equation}
\hat{\gamma_i}  = \frac{2}{h_i \Big(r_{i+1/2}^2-r_{i-1/2}^2 \Big)}  \Bigg[ \frac{r_{i+1/2}^3 - r_{i-1/2}^3}{3}  - r_{i-1/2} \frac{r_{i+1/2}^2 - r_{i-1/2}^2}{2} \Bigg]
\end{equation}

\begin{equation}
\hat{\gamma_i}  = \frac{\frac{2}{3}r_{i+1/2}^3 - r_{i-1/2} r_{i+1/2}^2 + \frac{1}{3}r_{i-1/2}^3}{h_i \Big(r_{i+1/2}^2-r_{i-1/2}^2 \Big)} \: .
\end{equation}

Note that 
\begin{equation}
h_i = r_{i+1/2}-r_{i-1/2} \: .
\end{equation}

To solve for $\gamma_i$, we need to make both sides of the following equation equivalent
\begin{equation}
\gamma_{i,2} \psi_{i+1/2} + (1-\gamma_{i,2}) \psi_{i-1/2} = \frac{\int_{r_{i-1/2}}^{r_{i+1/2}} dr \, r^2 \tilde{\psi}}{\int_{r_{i-1/2}}^{r_{i+1/2}} dr \, r^2}
\end{equation}
where 
\begin{equation}
\tilde{\psi} = \psi_{i+1/2} \Big( \frac{r-r_{i-1/2}}{h_i} \Big) +  \psi_{i-1/2} \Big( \frac{r_{i+1/2}-r}{h_i} \Big)  \: .
\end{equation}

This yields that
\begin{equation}
\gamma_{i,2} \psi_{i+1/2} + (1-\gamma_{i,2}) \psi_{i-1/2}  = \frac{1}{\Big(\frac{r_{i+1/2}^3-r_{i-1/2}^3}{3}\Big)} \int_{r_{i-1/2}}^{r_{i+1/2}} dr \, \Bigg[ \psi_{i+1/2} \Big( \frac{r^3-r_{i-1/2}r^2}{h_i} \Big) + \Big( \psi_{i-1/2} \frac{r_{i+1/2}r^2-r^3}{h_i} \Big) \Bigg]
\end{equation}
\begin{equation}
\gamma_{i,2} \psi_{i+1/2} + (1-\gamma_{i,2}) \psi_{i-1/2}  = \frac{3}{h_i \Big(r_{i+1/2}^3-r_{i-1/2}^3 \Big)}  \Bigg[ \Big(\psi_{i+1/2} - \psi_{i-1/2} \Big) \frac{r_{i+1/2}^4 - r_{i-1/2}^4}{4} \, + \, \Big(\psi_{i-1/2}r_{i+1/2} - \psi_{i+1/2}r_{i-1/2} \Big) \frac{r_{i+1/2}^3 - r_{i-1/2}^3}{3} \Bigg] \: .
\end{equation}

Now, by combining the terms that have a $\psi_{i+1/2}$, we can solve for $\gamma_{i,1}$ and get that 
\begin{equation}
\gamma_{i,2}  = \frac{3}{h_i \Big(r_{i+1/2}^3-r_{i-1/2}^3 \Big)}  \Bigg[ \frac{r_{i+1/2}^4 - r_{i-1/2}^4}{4}  - r_{i-1/2} \frac{r_{i+1/2}^3 - r_{i-1/2}^3}{3} \Bigg]
\end{equation}

\begin{equation}
\gamma_{i,2}  = \frac{\frac{3}{4}r_{i+1/2}^4 - r_{i-1/2} r_{i+1/2}^3 + \frac{1}{4}r_{i-1/2}^4}{h_i \Big(r_{i+1/2}^3-r_{i-1/2}^3 \Big)} \: .
\end{equation}


```python
import numpy as np
import numpy.polynomial.legendre as leg
import matplotlib.pyplot as plt
import copy

class sphere_mesh:
    def __init__(self, R=1., num_cells=2, num_angles=4):
        self.R = R
        self.num_cells = num_cells
        self.num_angles = num_angles
        
        self.mu, self.w = leg.leggauss(num_angles)
    
        self.mu_left  = np.zeros(num_angles); self.mu_right = np.zeros(num_angles)
        self.beta = np.zeros(num_angles)
    
        for m in range(num_angles):
            if m == 0:
                self.mu_left[m] = -1 
            else:
                self.mu_left[m] = self.mu_right[m-1]   
            self.mu_right[m] = self.mu_left[m] + self.w[m]
            self.beta[m] = (self.mu[m]-self.mu_left[m])/(self.mu_right[m]-self.mu_left[m])
            
        self.alpha_left = np.zeros(num_angles)
        self.alpha_right = np.zeros(num_angles)
        
        for m in range(num_angles):
            if m == 0:
                self.alpha_left[m] = 0.
            else:
                self.alpha_left[m] = self.alpha_right[m-1]
            self.alpha_right[m] = self.alpha_left[m] - 2*self.mu[m]*self.w[m]        
   
        self.h = np.ones(num_cells)*R/num_cells
        self.r_left = np.zeros(num_cells);   self.r_right = np.zeros(num_cells)
        self.r = np.zeros(num_cells)
    
        for i in range(num_cells):
            if i == 0:
                self.r_left[i] = 0 
            else:
                self.r_left[i] = self.r_right[i-1]
            self.r[i] = self.r_left[i] + self.h[i]/2.
            self.r_right[i] = self.r_left[i] + self.h[i]
            
        self.gammahat = np.zeros(num_cells)
        self.gamma = np.zeros(num_cells)
        self.A_left = np.zeros(num_cells)
        self.A_right = np.zeros(num_cells)
        self.V = np.zeros(num_cells)  
        
        r_left = self.r_left
        r_right = self.r_right
        h = self.h
        for i in range(num_cells):
            self.gammahat[i] = (2./3.*r_right[i]**3 - r_left[i]*r_right[i]**2 + 1./3.*r_left[i]**3) / (h[i]*(r_right[i]**2 - r_left[i]**2))
            self.gamma[i] = (3./4.*r_right[i]**4 - r_left[i]*r_right[i]**3 + 1./4.*r_left[i]**4) / (h[i]*(r_right[i]**3 - r_left[i]**3))
            self.A_left[i] = 4.*np.pi*r_left[i]**2
            self.A_right[i] = 4.*np.pi*r_right[i]**2
            self.V[i] = 4./3.*np.pi*(r_right[i]**3-r_left[i]**3)
```

```python
class boundary_condition:
    def __init__(self, mesh, type="vacuum", psi_out=0., phi=0., J_in=0., psi_1=0., value=0.):
        self.w = mesh.w
        self.mu = mesh.mu
        self.type = type
        self.psi_out = psi_out
        self.phi = phi
        self.value = value
        self.J_in = J_in
        
        self.psi_start = 0.
        self.psi_in = np.zeros(mesh.num_angles/2)
        w = self.w 
        mu = self.mu       
        
        if self.type == "partial_current":
            self.psi_start = -self.J_in / sum( w[0:len(w)/2] * mu[0:len(mu)/2] )
            self.psi_in = -self.J_in / sum( w[0:len(w)/2] * mu[0:len(mu)/2] ) * np.ones(mesh.num_angles/2)
        elif self.type == "flux":
            self.psi_start = self.phi / 2.
            self.psi_in = self.phi / 2. * np.ones(mesh.num_angles/2)
        elif self.type == "fixed_value":
            self.psi_start = self.value
            self.psi_in = self.value * np.ones(mesh.num_angles/2)
        elif self.type == "anisotropic_flux":
            if psi_1 > 1e-14:
                self.psi_in[0] = psi_1
            elif J_in > 1e-14:
                self.psi_in[0] = -self.J_in / (w[0] * mu[0])
            
            self.psi_start = self.psi_in[0] * (-1 - mu[1])/(mu[0] - mu[1])
        else: 
            print "\n\n\n\n** Error Incorrect Boundary Condition!"
            
```

```python
class physics: 
    def __init__(self, mesh, num_moments=1, sigma_t=0., sigma_s0=0., q0=0., sigma_s=None, q=None):
        self.num_moments = num_moments 
        self.phi = np.zeros((self.num_moments, mesh.num_cells)) 
        self.phi_left = np.zeros((self.num_moments, mesh.num_cells))
        self.phi_right = np.zeros((self.num_moments, mesh.num_cells)) 
        
        self.psi_start = np.zeros((mesh.num_cells)) 
        self.psi_start_left = np.zeros((mesh.num_cells))
        self.psi_start_right = np.zeros((mesh.num_cells))
        self.psihat_start_next = np.zeros((mesh.num_cells))
        
        self.psi = np.zeros((mesh.num_cells, mesh.num_angles))
        self.psihat = np.zeros((mesh.num_cells, mesh.num_angles))
        self.psi_left = np.zeros((mesh.num_cells, mesh.num_angles))
        self.psi_right = np.zeros((mesh.num_cells, mesh.num_angles))
        self.psihat_prev = np.zeros((mesh.num_cells, mesh.num_angles))
        self.psihat_next = np.zeros((mesh.num_cells, mesh.num_angles))  
        self.psi_boundary = np.zeros((mesh.num_angles))  
        
        self.J_left = np.zeros(mesh.num_cells)
        self.J_right = np.zeros(mesh.num_cells)
        self.J_left_minus = np.zeros(mesh.num_cells)
        self.J_left_plus = np.zeros(mesh.num_cells)
        self.J_right_minus = np.zeros(mesh.num_cells)
        self.J_right_plus = np.zeros(mesh.num_cells)
                
        self.sigma_t = sigma_t*np.ones(mesh.num_cells)  

        if sigma_s != None:
            self.sigma_s = sigma_s
        else:
            self.sigma_s = np.zeros((self.num_moments, mesh.num_cells))
            self.sigma_s[0,:] = sigma_s0*np.ones(mesh.num_cells) 
            
        if q != None:
            self.q = q
        else:
            self.q = np.zeros((self.num_moments, mesh.num_cells))
            self.q[0,:] = q0*np.ones(mesh.num_cells)  
        
        self.sigma_a = self.sigma_t - self.sigma_s[0,:]
        if num_moments == 1:
            self.sigma_tr = np.copy( self.sigma_t )
        else:
            self.sigma_tr = self.sigma_t - self.sigma_s[1,:]
            
        self.Q = (q0/2.)*np.ones((mesh.num_cells, mesh.num_angles))
        self.Q_start = (q0/2.)*np.ones((mesh.num_cells))
            
    def update():
        self.sigma_a = self.sigma_t - self.sigma_s[0,:]
        if self.num_moments == 1:
            self.sigma_tr = np.copy( self.sigma_t )
        else:
            self.sigma_tr = self.sigma_t - self.sigma_s[1,:]
            
        
```

The first step in solving the $S_N$ equation is to sweep for the starting direction flux $\mu = -1$. We start with the outermost cell (because we know what the incoming flux is) and sweep towards the innermost cells of the sphere. Thus we need to obtain an equation for $\psi_{i-1/2,1/2}$ in terms of $\psi_{i+1/2,1/2}$

\begin{equation}
- \psi_{i+1/2,1/2} + \psi_{i-1/2,1/2} + \sigma_t \bar{\psi}_{i,1/2} \Delta r = Q_{i,1/2} \Delta r
\end{equation}

\begin{equation}
- \psi_{i+1/2,1/2} + \psi_{i-1/2,1/2} + \frac{\sigma_t \Delta r}{2} (\psi_{i+1/2,1/2} + \psi_{i-1/2,1/2})= Q_{i,1/2} \Delta r
\end{equation}

\begin{equation}
\psi_{i-1/2,1/2}= \frac{\Big(1 - \frac{\sigma_t \Delta r}{2}\Big) \psi_{i+1/2,1/2} + Q_{i,1/2} \Delta r}{1 + \frac{\sigma_t \Delta r}{2}} \: .
\end{equation}
From here we can calculate angular derivate term (which is needed to solve the equation for the next direction, $\mu_1$),
\begin{equation}
\hat{\psi}_{i,1/2} = \gamma_{i,1} \psi_{i+1/2,1/2} + (1-\gamma_{i,1}) \psi_{i-1/2,1/2}  \: .
\end{equation}



```python
def sweep_startingflux(mesh, phys, BC, DEBUG):
    h = mesh.h
    I = mesh.num_cells
    gammahat = mesh.gammahat
    
    psi_right = phys.psi_start_right
    psi_left = phys.psi_start_left
    psihat_next = phys.psihat_start_next

    sigma_t = phys.sigma_t
    Q = phys.Q_start

    for i in reversed(range(I)):
        if i == I-1:
            psi_right[I-1] = BC.psi_start
        else:
            psi_right[i] = psi_left[i+1]
            
        psi_left[i] = ((1.-sigma_t[i]*h[i]/2.)*psi_right[i] + Q[i]*h[i]) / (1.+sigma_t[i]*h[i]/2.)
              
        psihat_next[i] = gammahat[i]*psi_right[i] + (1.-gammahat[i])*psi_left[i]

    phys.psi_start_left = psi_left        
    phys.psi_start_right = psi_right
    phys.psihat_start_next = psihat_next
        
    return phys
```

Once we know the starting flux in all spatial cells, we can sweep for the direction in the quadrature set $\mu_1$. Once again, we start with the outermost cell (because we know what the incoming flux is) and sweep towards the innermost cells of the sphere. We already know $\hat{\psi}_{i,m-1/2}$ from the previous sweep so we can derive an equation for $\psi_{i-1/2,m}$ in terms of $\psi_{i+1/2,m}$, 

\begin{equation}
\mu_m(A_{i+1/2} \psi_{i+1/2,m} - A_{i-1/2} \psi_{i-1/2,m} ) + \frac{1}{2}(A_{i+1/2} - A_{i-1/2}) \frac{(\alpha_{m+1/2} \hat{\psi}_{i,m+1/2} - \alpha_{m-1/2} \hat{\psi}_{i,m-1/2})}{w_m} + \sigma_t V_i \bar{\psi}_{i,m} = Q_{i,m} V_i
\end{equation}
where
\begin{equation}
\hat{\psi}_{i,m+1/2} = \frac{1}{\beta_m} \Bigg[ \gamma_{i,1} \psi_{i+1/2,m} + (1-\gamma_{i,1}) \psi_{i-1/2,m} - (1-\beta_m)\hat{\psi}_{i,m-1/2} \Bigg]
\end{equation}
\begin{equation}
\bar{\psi}_{i,m} = \gamma_{i,2} \psi_{i+1/2,m} + (1-\gamma_{i,2}) \psi_{i-1/2,m} \: .
\end{equation}

By expanding all terms we get that,
\begin{equation}
\mu_m A_{i+1/2} \psi_{i+1/2,m} - \mu_m A_{i-1/2} \psi_{i-1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} \gamma_{i,1} \psi_{i+1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\gamma_{i,1}) \psi_{i-1/2,m} - \\ \quad \quad \quad \quad \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m)\hat{\psi}_{i,m-1/2}  - \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m} \alpha_{m-1/2} \hat{\psi}_{i,m-1/2} + \sigma_t V_i \gamma_{i,2} \psi_{i+1/2,m} + \sigma_t V_i (1-\gamma_{i,2})  \psi_{i-1/2,m} = Q_{i,m} V_i \: .
\end{equation}

This leaves us with an equation with only one unknown $\psi_{i-1/2}$. By manipulating the equation above we get

\begin{equation}
 - \mu_m A_{i-1/2} \psi_{i-1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\gamma_{i,1}) \psi_{i-1/2,m} + \sigma_t V_i (1-\gamma_{i,2}) \psi_{i-1/2,m} = - \mu_m A_{i+1/2} \psi_{i+1/2,m} - \\ \quad \quad \quad \quad  \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} \gamma_{i,1} \psi_{i+1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m)\hat{\psi}_{i,m-1/2}  + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m} \alpha_{m-1/2} \hat{\psi}_{i,m-1/2} - \sigma_t V_i \gamma_{i,2} \psi_{i+1/2,m} + Q_{i,m} V_i
\end{equation}

\begin{equation}
\psi_{i-1/2,m} = \frac{\Bigg[- \mu_m A_{i+1/2} - \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} \gamma_{i,1} - \sigma_t V_i \gamma_{i,2} \Bigg]\psi_{i+1/2,m} + \Bigg[\frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m) + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m} \alpha_{m-1/2}\Bigg] \hat{\psi}_{i,m-1/2} + Q_{i,m} V_i}{- \mu_m A_{i-1/2} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\gamma_{i,1}) + \sigma_t V_i (1-\gamma_{i,2})} \: .
\end{equation}

Afterwards, we can solve for $\hat{\psi}_{i,m+1/2}$ by plugging in the appropriate values into
\begin{equation}
\hat{\psi}_{i,m+1/2} = \frac{1}{\beta_m} \Bigg[ \gamma_{i,1} \psi_{i+1/2,m} + (1-\gamma_{i,1}) \psi_{i-1/2,m} - (1-\beta_m)\hat{\psi}_{i,m-1/2} \Bigg] \: .
\end{equation}

Once sweep all the way to the first cell, we realize we have an overconstrianed system because $\psi_{3/2,m}$ is known and  $\psi_{1/2,m}$ (because we set it to be equal to the starting direction flux $\psi_{1/2,1/2}$). However, we can relax the condition that $\psi_{1,m}$ should be determine weighted-diamond difference, and instead use the balance equation to solve for $\psi_{1,m}$

\begin{equation}
\mu_m(A_{3/2} \psi_{3/2,m} - A_{1/2} \psi_{1/2,m} ) + \frac{1}{2}(A_{3/2} - A_{1/2}) \frac{(\alpha_{m+1/2} \hat{\psi}_{1,m+1/2} - \alpha_{m-1/2} \hat{\psi}_{1,m-1/2})}{w_m} + \sigma_t V_1 \psi_{1,m} = Q_{1,m} V_1
\end{equation}
where
\begin{equation}
\psi_{1/2,m}  = \psi_{1/2,1/2} 
\end{equation}
\begin{equation}
\hat{\psi}_{1,m+1/2} = \frac{1}{\beta_m} \Bigg[\hat{\psi}_{1,m} - (1-\beta_m)\hat{\psi}_{1,m-1/2} \Bigg]
\end{equation}
\begin{equation}
\hat{\psi}_{1,m} = \Bigg[ \frac{ \gamma_{1,i}\psi_{3/2,m} + (1-\gamma_{1,i}) \psi_{1/2,m} }{ \gamma_{2,i}\psi_{3/2,m} + (1-\gamma_{2,i}) \psi_{1/2,m} } \Bigg] \psi_{1,m}
\end{equation}
and 
\begin{equation}
A_{1/2} = 0 \: .
\end{equation}

This time since we know $\psi_{3/2,m}$ and $\psi_{1/2,m}$ for $\mu_m < 0$, we will instead solve for $\psi_{1,m}$. Thus, by manipulating the equation above we get

\begin{equation}
\mu_m A_{3/2} \psi_{3/2,m} + \frac{A_{3/2}}{2 w_m \beta_m} \alpha_{m+1/2} \Bigg[ \frac{ \gamma_{1,m}\psi_{3/2,m} + (1-\gamma_{1,m}) \psi_{1/2,m} }{ \gamma_{2,m}\psi_{3/2,m} + (1-\gamma_{2,m}) \psi_{1/2,m} } \Bigg] \psi_{1,m}  - \\ \quad \quad \quad \quad \frac{A_{3/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m)\hat{\psi}_{1,m-1/2}  - \frac{A_{3/2}}{2 w_m} \alpha_{m-1/2} \hat{\psi}_{1,m-1/2} + \sigma_t V_1 \psi_{1,m} = Q_{1,m} V_1
\end{equation}

\begin{equation}
\psi_{1,m} = \frac{ \Bigg[- \mu_m A_{3/2} \psi_{3/2,m} \Bigg] + \Bigg[ \frac{A_{3/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m) + \frac{A_{3/2}}{2 w_m} \alpha_{m-1/2} \Bigg] \hat{\psi}_{1,m-1/2} + Q_{1,m} V_1 }{\sigma_t V_1 + \frac{A_{3/2}}{2 w_m \beta_m} \alpha_{m+1/2} \Bigg[ \frac{ \gamma_{1,m}\psi_{3/2,m} + (1-\gamma_{1,m}) \psi_{1/2,m} }{ \gamma_{2,m}\psi_{3/2,m} + (1-\gamma_{2,m}) \psi_{1/2,m} } \Bigg]} \: .
\end{equation}

```python
def sweep_inward(mesh, phys, BC, DEBUG):

    I = mesh.num_cells
    N = mesh.num_angles
    mu = mesh.mu
    w = mesh.w
    A_left = mesh.A_left
    A_right = mesh.A_right
    V = mesh.V
    alpha_left = mesh.alpha_left
    alpha_right = mesh.alpha_right
    beta = mesh.beta
    gamma = mesh.gamma
    gammahat = mesh.gammahat
    
    psi = phys.psi
    psihat = phys.psihat
    psi_right = phys.psi_right
    psi_left = phys.psi_left
    psihat_prev = phys.psihat_prev
    psihat_next = phys.psihat_next

    sigma_t = phys.sigma_t
    Q = phys.Q
    
    for m in range(N/2):
        if DEBUG == True: print m
        for i in reversed(range(I)):
            if i == I-1:
                psi_right[I-1,m] = BC.psi_in[m]
            else:
                psi_right[i,m] = psi_left[i+1,m]
        
            if m == 0:
                psihat_prev[i,0] = phys.psihat_start_next[i]
            else:
                psihat_prev[i,m] = psihat_next[i,m-1]
                    
            if i == 0:
                psi_left[0,m] = phys.psi_start_left[0]
                
                gamma_ratio = (gammahat[i]*psi_right[i,m]+(1.-gammahat[i])*psi_left[i,m])/(gamma[i]*psi_right[i,m]+(1.-gamma[i])*psi_left[i,m])
                A = - mu[m]*A_right[i]*psi_right[i,m] + mu[m]*A_left[i]*psi_left[i,m]
                B = (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*(1.-beta[m]) + (A_right[i]-A_left[i])/(2.*w[m])*alpha_left[m]
                C = sigma_t[i]*V[i] + (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*gamma_ratio
            
                psi[i,m] = (A + B*psihat_prev[i,m] + Q[i,m]*V[i]) / C
                
                psihat[i,m] = gamma_ratio * psi[i,m]
                          
            else:
                A = - mu[m]*A_right[i] - (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*gammahat[i] - sigma_t[i]*V[i]*gamma[i]
                B = (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*(1.-beta[m]) + (A_right[i]-A_left[i])/(2*w[m])*alpha_left[m]
                C = - mu[m]*A_left[i] + (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*(1.-gammahat[i]) + sigma_t[i]*V[i]*(1.-gamma[i])      
            
                psi_left[i,m] = (A*psi_right[i,m] + B*psihat_prev[i,m] + Q[i,m]*V[i]) / C
                
                psi[i,m] = gamma[i]*psi_right[i,m] + (1.-gamma[i])*psi_left[i,m]
                
                psihat[i,m] = gammahat[i]*psi_right[i,m] + (1.-gammahat[i])*psi_left[i,m]
                
            psihat_next[i,m] = 1./beta[m]*(psihat[i,m] - (1.-beta[m])*psihat_prev[i,m])
            
            if DEBUG == True: print psi_left[i,m], psi[i,m], psi_right[i,m], psihat_prev[i,m], psihat_next[i,m]
                
        #print "psi[I-1,"+str(m)+"] = "+str(psi[I-1,m])
        #print "psi[0,"+str(m)+"] = "+str(psi[0,m])
        
    return phys
```

Once we know the flux in all spatial cells for directions $\mu<0$, we can sweep for the directions $\mu>0$. This time we start with the innermost cell in the sphere and sweep outwards. We set $\psi_{1/2,m}$ equal to the starting direction flux $\psi_{1/2,1/2}$. Also, we already know $\hat{\psi}_{i,m-1/2}$ from the previous sweep so we can derive an equation for $\psi_{i+1/2,m}$ in terms of $\psi_{i-1/2,m}$,

\begin{equation}
\mu_m A_{i+1/2} \psi_{i+1/2,m} - \mu_m A_{i-1/2} \psi_{i-1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} \gamma_{i,1} \psi_{i+1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\gamma_{i,1}) \psi_{i-1/2,m} - \\ \quad \quad \quad \quad \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m)\hat{\psi}_{i,m-1/2}  - \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m} \alpha_{m-1/2} \hat{\psi}_{i,m-1/2} + \sigma_t V_i \gamma_{i,2} \psi_{i+1/2,m} + \sigma_t V_i (1-\gamma_{i,2}) \psi_{i-1/2,m} = Q_{i,m} V_i \: .
\end{equation}

This leaves us with an equation with only one unknown $\psi_{i+1/2}$. By manipulating the equation above we get

\begin{equation}
\mu_m A_{i+1/2} \psi_{i+1/2,m} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} \gamma_{i,1} \psi_{i+1/2,m} + \sigma_t V_i \gamma_{i,2} \psi_{i+1/2,m} = \mu_m A_{i-1/2} \psi_{i-1/2,m} - \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\gamma_{i,1}) \psi_{i-1/2,m} - \\ \quad \quad \quad \quad \quad \quad \sigma_t V_i (1-\gamma_{i,2}) \psi_{i-1/2,m} +  \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m)\hat{\psi}_{i,m-1/2}  + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m} \alpha_{m-1/2} \hat{\psi}_{i,m-1/2} + Q_{i,m} V_i
\end{equation}

\begin{equation}
\psi_{i+1/2,m} = \frac{\Bigg[\mu_m A_{i-1/2} - \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\gamma_{i,1}) - \sigma_t V_i (1-\gamma_{i,2})  \Bigg]\psi_{i-1/2,m} + \Bigg[\frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} (1-\beta_m) + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m} \alpha_{m-1/2} \Bigg] \hat{\psi}_{i,m-1/2} + Q_{i,m} V_i}{\mu_m A_{i+1/2} + \frac{A_{i+1/2} - A_{i-1/2}}{2 w_m \beta_m} \alpha_{m+1/2} \gamma_{i,1} + \sigma_t V_i \gamma_{i,2} } \: .
\end{equation}

Afterwards, we can solve for $\hat{\psi}_{i,m+1/2}$ by plugging in the appropriate values into
\begin{equation}
\hat{\psi}_{i,m+1/2} = \frac{1}{\beta_m} \Bigg[ \gamma_{i,1} \psi_{i+1/2,m} + (1-\gamma_{i,1}) \psi_{i-1/2,m} - (1-\beta_m)\hat{\psi}_{i,m-1/2} \Bigg] \: .
\end{equation}

```python
def sweep_outward(mesh, phys, DEBUG):

    I = mesh.num_cells
    N = mesh.num_angles
    mu = mesh.mu
    w = mesh.w
    A_left = mesh.A_left
    A_right = mesh.A_right
    V = mesh.V
    alpha_left = mesh.alpha_left
    alpha_right = mesh.alpha_right
    beta = mesh.beta
    gamma = mesh.gamma
    gammahat = mesh.gammahat
    
    psi = phys.psi
    psihat = phys.psihat
    psi_right = phys.psi_right
    psi_left = phys.psi_left
    psihat_prev = phys.psihat_prev
    psihat_next = phys.psihat_next

    sigma_t = phys.sigma_t
    Q = phys.Q
    
    for m in range(N/2,N):
        if DEBUG == 1: print m
        for i in range(I):
            if i == 0:
                psi_left[0,m] = phys.psi_start_left[0]
            else:
                psi_left[i,m] = psi_right[i-1,m]
        
            if m == 0:
                psihat_prev[i,0] = psihat_start_next[i]
            else:
                psihat_prev[i,m] = psihat_next[i,m-1]
            
            A = mu[m]*A_left[i] - (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*(1.-gammahat[i]) - sigma_t[i]*V[i]*(1.-gamma[i]) 
            B = (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*(1.-beta[m]) + (A_right[i]-A_left[i])/(2.*w[m])*alpha_left[m]
            C = mu[m]*A_right[i] + (A_right[i]-A_left[i])/(2.*w[m]*beta[m])*alpha_right[m]*gammahat[i] + sigma_t[i]*V[i]*gamma[i]
            psi_right[i,m] = (A*psi_left[i,m] + B*psihat_prev[i,m] + Q[i,m]*V[i]) / C
            
            psi[i,m] = gamma[i]*psi_right[i,m] + (1.-gamma[i])*psi_left[i,m]
            psihat[i,m] = gammahat[i]*psi_right[i,m] + (1.-gammahat[i])*psi_left[i,m]
            psihat_next[i,m] = 1./beta[m]*(psihat[i,m] - (1.-beta[m])*psihat_prev[i,m])
            
            if DEBUG == True: print psi_left[i,m], psi_right[i,m]

        #print "psi[I-1,"+str(m)+"] = "+str(psi[I-1,m])
        #print "psi[0,"+str(m)+"] = "+str(psi[0,m])
                
    return phys
```

```python
def update_source(sigma_s, phi, q, mu):
    num_cells = len(phi[0])
    num_moments = max(len(sigma_s) , len(q))
    num_angles = len(mu)
    
    Q = np.zeros((num_cells, num_angles))
    for k in range(num_moments):
        coeffs = np.zeros(num_moments)
        coeffs[k] = 1.
        
        for m in range(num_angles):
            Q[:,m] += (2.*k+1.)/2. * (sigma_s[k]*phi[k] + q[k]) * leg.legval(mu[m],coeffs)      
    return Q   
            
```

```python
def compute_moments(mesh, phys):
    
    mu = mesh.mu
    w = mesh.w
    
    for k in range(phys.num_moments):
        for m in range(mesh.num_angles):
            # k is the moment, i is the cells, m is the direction
            # for phi the indices are [k,i]
            # for psi the indices are [i,m]
            coeffs = np.zeros(phys.num_moments)
            coeffs[k] = 1.
            #print phys.psi[:,m], leg.legval(mu[m],coeffs), w[m]
            phys.phi[k,:] += phys.psi[:,m]*leg.legval(mu[m],coeffs)*w[m]
            phys.phi_left[k,:] += phys.psi_left[:,m]*leg.legval(mu[m],coeffs)*w[m]
            phys.phi_right[k,:] += phys.psi_right[:,m]*leg.legval(mu[m],coeffs)*w[m]
    
    return phys
```

```python
def balance(mesh, phys, Q, table=True):
    
    w = mesh.w
    mu = mesh.mu
       
    for m in range(mesh.num_angles):
        if mesh.mu[m] < 0: 
            phys.J_left_minus[:] += - phys.psi_left[:,m] * w[m] * mu[m]
            phys.J_right_minus[:] += - phys.psi_right[:,m] * w[m] * mu[m]
        elif mesh.mu[m] > 0: 
            phys.J_left_plus[:] += phys.psi_left[:,m] * w[m] * mu[m]
            phys.J_right_plus[:] += phys.psi_right[:,m] * w[m] * mu[m]
        
    phys.J_left = phys.J_left_plus - phys.J_left_minus
    phys.J_right = phys.J_right_plus - phys.J_right_minus
    
    Q_ext = np.zeros(mesh.num_cells)
    for m in range(mesh.num_angles):
        Q_ext += Q[:,m] * w[m]
    
    I = mesh.num_cells
    phys.boundary_inflow = (mesh.A_right[I-1] * phys.J_right_minus[I-1]) 
    phys.boundary_outflow = (mesh.A_right[I-1] * phys.J_right_plus[I-1]) 
    phys.absorption = sum( phys.sigma_a * phys.phi[0,:] * mesh.V )
    phys.source_rate = sum( Q_ext * mesh.V ) 
    phys.balance = abs((phys.boundary_inflow + phys.source_rate - phys.absorption - phys.boundary_outflow)/(phys.boundary_inflow + phys.source_rate))
    
    if table==True: 
        print '\n---------------------------------'
        print 'inflow      = ', phys.boundary_inflow
        print 'outflow     = ', phys.boundary_outflow
        print 'absorption  = ', phys.absorption
        print 'source rate = ', phys.source_rate
        print '---------------------------------'
        print 'balance     = ', phys.balance
        print '---------------------------------\n'

    return phys
```

0th moment equation:

\begin{equation}
\sum_{m=1}^N w_m \Bigg[ \frac{\mu_m}{r^2} \frac{\partial (r^2 \psi_m)}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2} - \alpha_{m-1/2} \psi_{m-1/2})}{w_m} + \sigma_t \psi_m = \sum_{k=0}^K \frac{2k+1}{2} (\sigma_k \phi_k + q_k) P_k(\mu_m) \Bigg]
\end{equation}

\begin{equation}
 \frac{1}{r^2} \frac{\partial (r^2 J)}{\partial r} \: + \: \sum_{m=1}^N  \Big[ \frac{1}{r} (\alpha_{m+1/2} \psi_{m+1/2} - \alpha_{m-1/2} \psi_{m-1/2}) \Big] \: + \:  \sigma_t \phi  \: = \:  \sum_{m=1}^N w_m P_0(\mu_m) \Big[ \sum_{k=0}^K \frac{2k+1}{2} (\sigma_k \phi_k + q_k) P_k(\mu_m) \Big]
\end{equation}

\begin{equation}
 \frac{1}{r^2} \frac{\partial (r^2 J)}{\partial r} \: + \:  \frac{1}{r} (\alpha_{N+1/2} \psi_{N+1/2} - \alpha_{1/2} \psi_{1/2}) \: + \:  \sigma_t \phi  \: = \:    \sum_{k=0}^K \frac{2 \delta_{0,k}}{2k+1} \frac{2k+1}{2} (\sigma_k \phi_k + q_k) 
\end{equation}

\begin{equation}
\alpha_{1/2} = \alpha_{N+1/2} = 0
\end{equation}

\begin{equation}
 \frac{1}{r^2} \frac{\partial (r^2 J)}{\partial r} + \sigma_t \phi  = \sigma_s \phi + q 
\end{equation}

\begin{equation}
\boxed{ \frac{1}{r^2} \frac{\partial (r^2 J)}{\partial r} + \sigma_a \phi  = q }
\end{equation}

1st moment equation:

\begin{equation}
\sum_{m=1}^N w_m \mu_m  \Bigg[ \frac{\mu_m}{r^2} \frac{\partial (r^2 \psi_m)}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2} - \alpha_{m-1/2} \psi_{m-1/2})}{w_m} + \sigma_t \psi = \sum_{k=0}^K \frac{2k+1}{2} (\sigma_k \phi_k + q_k) P_k(\mu_m) \Bigg]
\end{equation}

\begin{equation}
 \frac{1}{r^2} \frac{\partial (r^2 \vec{E} \phi)}{\partial r} \: + \: \sum_{m=1}^N  \Big[ \frac{\mu_m}{r} (\alpha_{m+1/2} \psi_{m+1/2} - \alpha_{m-1/2} \psi_{m-1/2}) \Big] \: + \:  \sigma_t J  \: = \:  \sum_{m=1}^N w_m P_1(\mu_m) \Big[ \sum_{k=0}^K \frac{2k+1}{2} (\sigma_k \phi_k + q_k) P_k(\mu_m) \Big]
\end{equation}

\begin{equation}
 \frac{1}{3r^2} \frac{\partial (r^2 \phi)}{\partial r} \: + \: \sum_{m=1}^N  \Big[ \frac{\mu_m}{r} (\alpha_{m+1/2} \psi_{m+1/2} - \alpha_{m-1/2} \psi_{m-1/2}) \Big] \: + \:  \sigma_t J  \: = \:  \sigma_{s,1} J + q_1 
\end{equation}

\begin{equation}
\psi_{m} = \frac{1}{2}\phi + \frac{3}{2} \mu_m J
\end{equation}

\begin{equation}
\sigma_{tr} = \sigma_t -  \sigma_{s,1}
\end{equation}

\begin{equation}
 \frac{1}{3} \frac{\partial \phi}{\partial r} \: + \:  \frac{2}{3r} \phi \: + \: \sum_{m=1}^N  \Big\{ \frac{\mu_m}{r} \Big[ \alpha_{m+1/2} (\frac{1}{2}\phi + \frac{3}{2} \mu_{m+1/2} J) - \alpha_{m-1/2} (\frac{1}{2}\phi + \frac{3}{2} \mu_{m-1/2} J) \Big] \Big\} \: + \:  \sigma_{tr} J \: = \: q_1 
\end{equation}

\begin{equation}
 \frac{1}{3} \frac{\partial \phi}{\partial r} \: + \:  \frac{2}{3r} \phi  \:  + \: \sum_{m=1}^N \frac{\mu_m}{r} (\frac{1}{2}\phi) (\alpha_{m+1/2} \: - \: \alpha_{m-1/2}) \: - \: \sum_{m=1}^N  \frac{\mu_m}{r} (\frac{3}{2} \mu_{m+1/2} J) (\alpha_{m+1/2}  - \alpha_{m-1/2}) \: + \:  \sigma_{tr} J \: = \: q_1 
\end{equation}

\begin{equation}
 \frac{1}{3} \frac{\partial \phi}{\partial r} \: + \:  \frac{2}{3r} \phi \:  + \: \sum_{m=1}^N \frac{\mu_m}{r} (\frac{1}{2}\phi) (-2\mu_m w_m) \: - \: \sum_{m=1}^N  \frac{\mu_m}{r} (\frac{3}{2} \mu_{m+1/2} J) (-2 \mu_m w_m) \: + \:  \sigma_{tr} J \: = \: q_1 
\end{equation}

\begin{equation}
 \frac{1}{3} \frac{\partial \phi}{\partial r} \: + \:  \frac{2}{3r} \phi \:  - \: \frac{2}{3r} \phi  \: + \: 3J \sum_{m=1}^N  \frac{\mu_m^2}{r} \mu_{m+1/2} w_m \: + \:  \sigma_{tr} J \: = \: q_1 
\end{equation}

\begin{equation}
\mu_{m+1/2} = \mu_m + w_m (1-\beta_m)  
\end{equation}

\begin{equation}
 \frac{1}{3} \frac{\partial \phi}{\partial r}  \: + \: \frac{3 J}{r}  \sum_{m=1}^N  \mu_m^3 w_m \: + \: \frac{3 J}{r}  \sum_{m=1}^N  \mu_m^2 w_m^2 (1-\beta_m) \: + \: \sigma_{tr} J \: = \: q_1 
\end{equation}

Note that $\mu_m$ and $\beta_m$ are both odd functions, and $w_m$ is an even function. Also, the integral of an odd function is zero, thus

\begin{equation}
\frac{3 J}{r}  \sum_{m=1}^N  \mu_m^3 w_m \:= \: \frac{3 J}{r}  \sum_{m=1}^N  \mu_m^2 w_m^2 (1-\beta_m) \:= \: 0
\end{equation}

and thus the first moment equation is 

\begin{equation}
\boxed{ \frac{1}{3} \frac{\partial \phi}{\partial r}  \: + \: \sigma_{tr} J \: = \: q_1 }
\end{equation}

In order to derive the $0^{th}$ and $1^{st}$ moment equation for DSA, we must first derive the transport equation for the error in iteration $\ell+1/2$. We start with the spatially-analytic transport equation with isotropic scattering 
\begin{equation}
\frac{\mu_m}{r^2} \frac{\partial (r^2 \psi_m)}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2} - \alpha_{m-1/2} \psi_{m-1/2})}{w_m} + \sigma_t \psi_m = \frac{1}{2} \sigma_{s} \phi + \sum_{k=0}^K \frac{2k +1}{2} q_k \: .
\end{equation}
Then we express the transport equation for source iteration $\ell+1/2$, which is 
\begin{equation}
\frac{\mu_m}{r^2} \frac{\partial (r^2 \psi_m^{\ell+1/2})}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2}^{\ell+1/2} - \alpha_{m-1/2} \psi_{m-1/2}^{\ell+1/2})}{w_m} + \sigma_t \psi_m^{\ell+1/2} = \frac{1}{2} \sigma_{s} \phi^{\ell} + \sum_{k=0}^K \frac{2k +1}{2} q_k  \: .
\end{equation}
By subtracting the equation for $\ell+1/2$ from the analytic equation we get a transport equation for the error
\begin{equation}
\frac{\mu_m}{r^2} \frac{\partial (r^2 \delta \psi_m^{\ell+1/2})}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2}^{\ell+1/2} - \alpha_{m-1/2} \delta \psi_{m-1/2}^{\ell+1/2})}{w_m} + \sigma_t \delta \psi_m^{\ell+1/2} = \frac{1}{2} \sigma_{s} \delta \phi^{\ell}   \: .
\end{equation}
where 
\begin{equation}
\delta \psi^\ell = \psi - \psi^\ell 
\end{equation}

By subtracting $\frac{1}{2} \sigma_{s} \delta \phi^{\ell+1/2}$ from both sides we get
\begin{equation}
\frac{\mu_m}{r^2} \frac{\partial (r^2 \delta \psi_m^{\ell+1/2})}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2}^{\ell+1/2} - \alpha_{m-1/2} \delta \psi_{m-1/2}^{\ell+1/2})}{w_m} + \sigma_t \delta \psi_m^{\ell+1/2} - \frac{1}{2} \sigma_s \delta \phi^{\ell+1/2} = \frac{1}{2} \sigma_{s} (\delta \phi^{\ell} - \delta \phi^{\ell+1/2}) 
\end{equation}

\begin{equation}
\frac{\mu_m}{r^2} \frac{\partial (r^2 \delta \psi_m^{\ell+1/2})}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2}^{\ell+1/2} - \alpha_{m-1/2} \delta \psi_{m-1/2}^{\ell+1/2})}{w_m} + \sigma_t \delta \psi_m^{\ell+1/2} - \frac{1}{2} \sigma_s \delta \phi^{\ell+1/2} = \frac{1}{2} \sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell}) \: .
\end{equation}

\begin{equation}
\frac{\mu_m}{r^2} \frac{\partial (r^2 \delta \psi_m^{\ell+1/2})}{\partial r} + \frac{1}{r} \frac{(\alpha_{m+1/2} \psi_{m+1/2}^{\ell+1/2} - \alpha_{m-1/2} \delta \psi_{m-1/2}^{\ell+1/2})}{w_m} + \sigma_t \delta \psi_m^{\ell+1/2} - \frac{1}{2} \sigma_s \delta \phi^{\ell+1/2} = \frac{1}{2} \sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell}) \: .
\end{equation}


Notice how the equation above for the error is just like the spatially-analytic tranpsort equation except the source term has no higher moments. Thus, the $0^{th}$ and $1^{st}$ moment equation for the error are same as for the angular flux, except that $q_0=\sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell})$ and $q_1=0$:
\begin{equation}
\boxed{ \frac{1}{r^2} \frac{\partial (r^2 \delta J^{\ell+1/2})}{\partial r} + \sigma_a \delta \phi^{\ell+1/2}  = \sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell})  }
\end{equation}
\begin{equation}
\boxed{ \frac{1}{3} \frac{\partial \delta \phi^{\ell+1/2}}{\partial r}  \: + \: \sigma_{tr} \delta J^{\ell+1/2} \: = \: 0 }
\end{equation}

The discrete form of the $0^th$ moment equation can be derived for cell $i$ integrating over volume element $i$:

\begin{equation}
(A_{i+1/2} \delta J_{i+1/2}^{\ell+1/2} - A_{i-1/2} \delta J_{i-1/2}^{\ell+1/2}) + \sigma_a [\gamma_i \delta \phi_{i+1/2}^{\ell+1/2} + (1-\gamma_i)\delta \phi_{i-1/2}^{\ell+1/2}] V_i = R_i^{\ell+1/2} V_i
\end{equation}

In addition, for cell $i+1$ the discrete 0th moment equation is 

\begin{equation}
(A_{i+3/2} \delta J_{i+3/2}^{\ell+1/2} - A_{i+1/2} \delta J_{i+1/2}^{\ell+1/2}) + \sigma_a [\gamma_{i+1} \delta \phi_{i+3/2}^{\ell+1/2} + (1-\gamma_{i+1})\delta \phi_{i+1/2}^{\ell+1/2}] V_{i+1} = R_{i+1}^{\ell+1/2} V_{i+1}
\end{equation}

by combining both of these equation we get that 

\begin{equation}
(A_{i+3/2} \delta J_{i+3/2}^{\ell+1/2} - A_{i-1/2} \delta J_{i-1/2}^{\ell+1/2}) + \sigma_a [\gamma_i \delta \phi_{i+1/2}^{\ell+1/2} + (1-\gamma_i) \delta \phi_{i-1/2}^{\ell+1/2}] V_i + \sigma_a [\gamma_{i+1} \delta \phi_{i+3/2}^{\ell+1/2} + (1-\gamma_{i+1}) \delta \phi_{i+1/2}^{\ell+1/2}] V_{i+1} = R_i^{\ell+1/2} V_i + R_{i+1}^{\ell+1/2} V_{i+1} 
\end{equation}

However we do not have an equation for $J$ at the cell edges, but we can make the approximation that 
\begin{equation}
A_{i+3/2} \delta J_{i+3/2}^{\ell+1/2} - A_{i-1/2} \delta J_{i-1/2}^{\ell+1/2} \approx 2(A_{i+1} \delta J_{i+1}^{\ell+1/2} - A_i \delta J_i^{\ell+1/2})
\end{equation}

and 

\begin{equation}
 \frac{1}{3} (\delta \phi_{i+1/2}^{\ell+1/2} - \delta \phi_{i-1/2}^{\ell+1/2})  \: + \: \sigma_{tr} \delta J_i^{\ell+1/2} h_i \: = \: 0 
\end{equation}
\begin{equation}
 \frac{1}{3} (\delta \phi_{i+3/2}^{\ell+1/2} - \delta \phi_{i+1/2}^{\ell+1/2})  \: + \: \sigma_{tr} \delta J_{i+1}^{\ell+1/2}  h_{i+1} \: = \: 0 
\end{equation}

This gives us a discrete form of the diffusion equation,
\begin{equation}
- \frac{2 A_{i+1} (\delta \phi_{i+3/2}^{\ell+1/2} - \delta \phi_{i+1/2}^{\ell+1/2})}{3 \sigma_{tr} h_{i+1}} + \frac{2 A_i (\delta \phi_{i+1/2}^{\ell+1/2} - \delta \phi_{i-1/2}^{\ell+1/2})}{3 \sigma_{tr} h_i} + \sigma_a [\gamma_i \delta \phi_{i+1/2}^{\ell+1/2} + (1-\gamma_i)\delta \phi_{i-1/2}^{\ell+1/2}] V_i + \sigma_a [\gamma_{i+1} \delta \phi_{i+3/2}^{\ell+1/2} + (1-\gamma_{i+1})\delta \phi_{i+1/2}^{\ell+1/2}] V_{i+1} = R_i^{\ell+1/2} V_i + R_{i+1}^{\ell+1/2} V_{i+1} 
\end{equation}

\begin{equation}
\Big\{- \frac{2 A_i }{3 \sigma_{tr} h_i} + \sigma_a(1-\gamma_i)V_i\Big\} \delta \phi_{i-1/2}^{\ell+1/2} + \Big\{ \frac{2 A_{i+1}}{3 \sigma_{tr} h_{i+1}} + \frac{2 A_i }{3 \sigma_{tr} h_i} + \sigma_a [\gamma_i V_i + (1-\gamma_{i+1}) V_{i+1}] \Big\} \delta \phi_{i+1/2}^{\ell+1/2}  + \Big\{- \frac{2 A_{i+1}}{3 \sigma_{tr} h_{i+1}} + \sigma_a \gamma_{i+1} V_{i+1}\Big\} \delta \phi_{i+3/2}^{\ell+1/2} = R_i^{\ell+1/2} V_i + R_{i+1}^{\ell+1/2} V_{i+1} 
\end{equation}
where
\begin{equation}
R^{\ell+1/2} = \sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell}) 
\end{equation}


The discrete form of the 0th moment equation can be derived for cell $1$ integrating over volume element $1$:

\begin{equation}
(A_{3/2} \delta J_{3/2}^{\ell+1/2} - A_{1/2} \delta J_{1/2}^{\ell+1/2}) + \sigma_a \big[\gamma_1 \delta \phi_{3/2}^{\ell+1/2} + (1-\gamma_1)\delta \phi_{1/2}^{\ell+1/2}\big] V_1 = R_1^{\ell+1/2} V_1
\end{equation}

but we know that $A_{1/2} = 0$ thus

\begin{equation}
(A_{3/2} \delta J_{3/2}^{\ell+1/2} - A_{1/2} \delta J_{1/2}^{\ell+1/2}) = A_{3/2} \delta J_{3/2}^{\ell+1/2} \: .
\end{equation}

In addition, since we don't know what $J_{3/2}$ we can make the assumption that 

\begin{equation}
A_{3/2} \delta J_{3/2}^{\ell+1/2} = 2 A_1 \delta J_1^{\ell+1/2}
\end{equation}

and 

\begin{equation}
 \frac{1}{3} (\delta \phi_{3/2}^{\ell+1/2} - \delta \phi_{1/2}^{\ell+1/2})  \: + \: \sigma_{tr} \delta J_1^{\ell+1/2} h_1 \: = \: 0 \: .
\end{equation}

This gives us a discrete form of the diffusion equation,
\begin{equation}
- \frac{2 A_1 (\delta \phi_{3/2}^{\ell+1/2} - \delta \phi_{1/2}^{\ell+1/2})}{3 \sigma_{tr} h_1} + \sigma_a [\gamma_1 \delta \phi_{3/2}^{\ell+1/2} + (1-\gamma_1)\delta \phi_{1/2}^{\ell+1/2}] V_1 = R_1^{\ell+1/2} V_1  
\end{equation}

\begin{equation}
\Big\{\frac{2 A_1}{3 \sigma_{tr} h_1} + \sigma_a(1-\gamma_1)V_1\Big\} \delta \phi_{1/2}^{\ell+1/2} + \Big\{- \frac{2 A_1}{3 \sigma_{tr} h_1} + \sigma_a \gamma_1 V_1\Big\} \delta \phi_{3/2}^{\ell+1/2} = R_1^{\ell+1/2} V_1 
\end{equation}
where
\begin{equation}
R^{\ell+1/2} = \sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell}) 
\end{equation}


The discrete form of the 0th moment equation can be derived for cell $N$ integrating over volume element $N$:

\begin{equation}
(A_{N+1/2} \delta J_{N+1/2}^{\ell+1/2} - A_{N-1/2} \delta J_{N-1/2}^{\ell+1/2}) + \sigma_a \big[\gamma_1 \delta \phi_{N+1/2}^{\ell+1/2} + (1-\gamma_1)\delta \phi_{N-1/2}^{\ell+1/2}\big] V_1 = R_1^{\ell+1/2} V_1
\end{equation}

Since we don't know what $J_{N-1/2}$ we can make the assumption that 

\begin{equation}
A_{N+1/2} \delta J_{N+1/2}^{\ell+1/2} - A_{N-1/2} \delta J_{N-1/2}^{\ell+1/2} \: = \: 2(A_{N+1/2} \delta J_{N+1/2}^{\ell+1/2} -  A_{N} \delta J_N^{\ell+1/2})
\end{equation}

and after plugging in the boundary condition this is equivalent to 

\begin{equation}
2(A_{N+1/2} \delta J_{N+1/2}^{\ell+1/2} -  A_{N} \delta J_N^{\ell+1/2}) \: = \: 2 A_{N+1/2} \delta  \phi_{N+1/2}^{\ell+1/2} \langle \mu \rangle - 2 A_{N} \delta J_N^{\ell+1/2} \: .
\end{equation}

This gives us a discrete form of the diffusion equation,
\begin{equation}
2 A_{N+1/2} \delta  \phi_{N+1/2}^{\ell+1/2} \langle \mu \rangle + \frac{2 A_N (\delta \phi_{N+1/2}^{\ell+1/2} - \delta \phi_{N-1/2}^{\ell+1/2})}{3 \sigma_{tr} h_N} + \sigma_a [\gamma_N \delta \phi_{N+1/2}^{\ell+1/2} + (1-\gamma_N)\delta \phi_{N-1/2}^{\ell+1/2}] V_N = R_N^{\ell+1/2} V_N  
\end{equation}

\begin{equation}
\Big\{- \frac{2 A_N}{3 \sigma_{tr} h_N} + \sigma_a(1-\gamma_N)V_N\Big\} \delta \phi_{N-1/2}^{\ell+1/2} + \Big\{2 A_{N+1/2} \langle \mu \rangle + \frac{2 A_N}{3 \sigma_{tr} h_N} + \sigma_a \gamma_N V_N\Big\} \delta \phi_{N+1/2}^{\ell+1/2} = R_N^{\ell+1/2} V_N 
\end{equation}
where
\begin{equation}
R^{\ell+1/2} = \sigma_{s} (\phi^{\ell+1/2} - \phi^{\ell}) 
\end{equation}

```python
def DSA(mesh, phys, phys_old, DEBUG):
    A = np.matrix( np.zeros((mesh.num_cells + 1, mesh.num_cells + 1)) )
    b = np.matrix( np.zeros((mesh.num_cells + 1, 1)) )
    I = mesh.num_cells
    
    mesh.A_avg = [(mesh.A_left[i] + mesh.A_right[i])/ 2. for i in range(I)]
    mu_avg = 0
    for m in range(mesh.num_angles):
        if mesh.mu[m] > 0:
            mu_avg += mesh.mu[m]*mesh.w[m] 
        
    # reflective boundary
    A[0, 0] =  2.*mesh.A_avg[0]/(3.*phys.sigma_tr[0]*mesh.h[0]) + \
               phys.sigma_a[0]*(1-mesh.gamma[0])*mesh.V[0]
    A[0, 1] =  -2.*mesh.A_avg[0]/(3.*phys.sigma_tr[0]*mesh.h[0]) + \
               phys.sigma_a[0]*mesh.gamma[0]*mesh.V[0]
    b[0, 0] =  phys.sigma_s[0,0]*(phys.phi[0,0] - phys_old.phi[0,0])*mesh.V[0] 
    
    for i in range(I-1):
        # the indices for A and b represent edges
        # the indices for everything else represent cells
        A[i+1,   i]  =  -2.*mesh.A_avg[i]/(3.*phys.sigma_tr[i]*mesh.h[i]) + \
                        phys.sigma_a[i]*(1.-mesh.gamma[i])*mesh.V[i]
                
        A[i+1, i+1]  =  2.*mesh.A_avg[i+1]/(3.*phys.sigma_tr[i+1]*mesh.h[i+1]) + \
                        2.*mesh.A_avg[i]/(3.*phys.sigma_tr[i]*mesh.h[i]) + \
                        (phys.sigma_t[i] - phys.sigma_s[0,i])*mesh.gamma[i]*mesh.V[i] + \
                        (phys.sigma_t[i+1] - phys.sigma_s[0,i+1])*(1.-mesh.gamma[i+1])*mesh.V[i+1]
        
        A[i+1, i+2]  =  -2.*mesh.A_avg[i+1]/(3.*phys.sigma_tr[i+1]*mesh.h[i+1]) + \
                        phys.sigma_a[i+1]*mesh.gamma[i+1]*mesh.V[i+1]
       
        b[i+1,   0]  =  phys.sigma_s[0,i]*(phys.phi[0,i] - phys_old.phi[0,i])*mesh.V[i] + \
                        phys.sigma_s[0,i+1]*(phys.phi[0,i+1] - phys_old.phi[0,i+1])*mesh.V[i+1]
            
    # vacuum boundary
    A[I, I-1] =  -2.*mesh.A_avg[I-1]/(3.*phys.sigma_tr[I-1]*mesh.h[I-1]) + \
                 phys.sigma_a[I-1]*(1-mesh.gamma[I-1])*mesh.V[I-1]
    A[I, I]   =  2.*mesh.A_right[I-1]*mu_avg + 2.*mesh.A_avg[I-1]/(3.*phys.sigma_tr[I-1]*mesh.h[I-1]) + \
                 phys.sigma_a[I-1]*mesh.gamma[I-1]*mesh.V[I-1]
    b[I, 0]   =  phys.sigma_s[0,I-1]*(phys.phi[0,I-1] - phys_old.phi[0,I-1])*mesh.V[I-1] 
              
    if DEBUG == True:
        print "phi", phys.phi
        print "phi_old", phys_old.phi
        print "A", A
        print "b", b
        
    x = np.linalg.solve(A, b).transpose()
    
    phys.phi_left_error = np.zeros(I)
    phys.phi_right_error = np.zeros(I)
    phys.phi_error = np.zeros(I)
    phys.J_error = np.zeros(I)
    
    for i in range(I):
        phys.phi_left_error[i] = x[0,i]
        phys.phi_right_error[i] = x[0,i+1]

    for i in range(I):
        phys.phi_error[i] = mesh.gamma[i]*phys.phi_right_error[i] + (1.-mesh.gamma[i])*phys.phi_left_error[i]
        phys.J_error[i] = -1./(3.*phys.sigma_tr[i]*mesh.h[i])*(phys.phi_right_error[i] - phys.phi_left_error[i])
        
        phys.phi[0,i] += phys.phi_error[i]
        
        if phys.num_moments != 1: 
            phys.phi[1,i] += phys.J_error[i]
        
        for m in range(mesh.num_angles):
            phys.psi[i,m] += 0.5*phys.phi_error[i] + 1.5*phys.J_error[i]*mesh.mu[m]
            
    for m in range(mesh.num_angles):
        if mesh.mu[m] > 0:
            phys.psi_right[I-1,m] += 0.5*phys.phi_right_error[I-1] + 1.5*phys.phi_right_error[I-1]*mu_avg*mesh.mu[m]      
            
    return phys
```

```python
import matplotlib.pyplot as plt
import copy

def source_iteration(mesh, phys, BC, tol=1e-8, max_its=100, include_DSA=False, DEBUG=False):

    phys_old = copy.deepcopy(phys)
    phys_reset = copy.deepcopy(phys)
    mu = mesh.mu
    w = mesh.w
    
    for l in range(max_its):
        
        phys_old_old = copy.deepcopy(phys_old) #copy previous "old" iteration values  
        phys_old = copy.deepcopy(phys) #copy previous iteration values            
        phys = copy.deepcopy(phys_reset) #reset all values in class phys to original value
        
        phys.Q = update_source(phys.sigma_s, phys_old.phi, phys.q, mu)

        phys.Q_start = phys.Q[:,0]*(mu[1]+1)/(mu[1]-mu[0]) - phys.Q[:,1]*(1+mu[0])/(mu[1]-mu[0])       
    
        # sweep through spherical geometry
        phys = sweep_startingflux(mesh, phys, BC, DEBUG) #for mu = -1
        phys = sweep_inward(mesh, phys, BC, DEBUG) #for mu < 0
        phys = sweep_outward(mesh, phys, DEBUG) #for mu > 0
        
        phys = compute_moments(mesh, phys)
        
        if include_DSA == True: 
            phys = DSA(mesh, phys, phys_old, DEBUG)
            
        phys.psi_boundary = np.copy( phys.psi_right[mesh.num_cells-1,:] )
                   
        delta_phi = abs((phys.phi[0,:] - phys_old.phi[0,:])/phys.phi[0,:])
        spectral_rad = sum(abs(phys.phi[0,:] - phys_old.phi[0,:])) / sum(abs(phys_old.phi[0,:] - phys_old_old.phi[0,:]))
        rel_err = delta_phi / (1 - spectral_rad) 
        
        if max(delta_phi) < tol*(1-spectral_rad): break
            
        print "\n\n>> ITERATION # ", l
        phys = balance(mesh, phys, phys_reset.Q)
            
    return phys
```

In order for the partial current to be equal to 1, 
\begin{equation}
J^- = - \sum_{m=1}^{M/2} \psi_m \mu_m w_m = 1 \: ,
\end{equation}
but only $\psi_1$ is non-zero, therefore 
\begin{equation}
\psi(R_o, \mu) = - \frac{\delta(\mu - \mu_1)}{\mu_1 w_1} \: .
\end{equation}

The asymtotic transport equation for a strong isotropic scattering slab
\begin{equation}
\mu \frac{\partial \psi}{\partial x} + (\sigma_s + \epsilon \sigma_a) \psi(x, \mu) = \frac{1}{2}\sigma_s \phi(x) 
\end{equation}
where 
\begin{equation}
\epsilon \ll 1
\end{equation}

with the boundary conditions
\begin{equation}
\psi_u(0, \mu) = \psi_u(0, -\mu) 
\end{equation}
and
\begin{equation}
\psi(R_o, \mu) = - \frac{\delta(\mu - \mu_1)}{\mu_1 w_1} \: .
\end{equation}

The $0^{th}$ order equation is 
\begin{equation}
\mu \frac{\partial \psi}{\partial x} + \sigma_s \psi(x, \mu) = \frac{1}{2}\sigma_s \phi(x) \: .
\end{equation}

This equation can be split up into two equations. One for the uncollided flux and another for the collided flux:
\begin{equation}
\mu \frac{\partial \psi_u}{\partial x} + \sigma_s \psi_u(x, \mu) = 0 
\end{equation}
\begin{equation}
\mu \frac{\partial \psi_c}{\partial x} + \sigma_s \psi_c(x, \mu) = \frac{1}{2}\sigma_s \phi_c(x) + \frac{1}{2}\sigma_s \phi_u(x) \: .
\end{equation}

We can easily solve the equation for the uncollided flux for $\mu < 0$, 
\begin{equation}
\psi_u(x, \mu) = \psi_u(R_o, \mu) \, \text{exp}\Big[\frac{\sigma_s}{\mu}(R_o-x)\Big]
\end{equation}
\begin{equation}
\psi_u(x, \mu) = - \frac{\delta(\mu - \mu_1)}{\mu_1 w_1} \text{exp}\Big[\frac{\sigma_s}{\mu}(R_o-x)\Big] \quad \text{for} \: \mu < 0 
\end{equation}
and for $\mu > 0$,
\begin{equation}
\psi_u(0, \mu) = \psi_u(0, -\mu) 
\end{equation}

\begin{equation}
\psi_u(x, \mu) = - \frac{\delta(\mu + \mu_1)}{\mu_1 w_1} \text{exp}\Big[-\frac{\sigma_s}{\mu}(R_o-0)\Big] \text{exp}\Big[- \frac{\sigma_s}{\mu}x\Big] 
\end{equation}

\begin{equation}

```python
###########################################################################
mesh_b = sphere_mesh(R=1., num_cells=50, num_angles=16)
hat = physics(mesh_b, sigma_t=3., sigma_s0=2., q0=0.)
BC_b = boundary_condition(mesh_b, type="fixed_value", value=1.)

hat = source_iteration(mesh_b, hat, BC_b, include_DSA=True, tol=1e-13,  max_its=500, DEBUG=False)
###########################################################################

adjoint = copy.deepcopy(hat)
for m in range(mesh_b.num_angles):
    adjoint.psi_boundary[m] = hat.psi_boundary[mesh_b.num_angles-1-m]
    for i in range(mesh_b.num_cells):
        adjoint.psi[i,m] = hat.psi[i, mesh_b.num_angles-1-m]

#plt.plot(mesh_b.r, adjoint.phi[0,:])
#plt.xlabel('$r$ $[cm]$', fontsize=20)
#plt.ylabel('$\phi$ $[p/cm^2-s]$', fontsize=20)
#plt.show()
        
R_adjoint = 0.
for m in range(mesh_b.num_angles):
    if mesh_b.mu[m] < 0:
        R_adjoint -= forward.psi_boundary[m]*adjoint.psi_boundary[m]*mesh_b.mu[m]*mesh_b.w[m]*(4.*np.pi*mesh_b.R**2)
        
print R_adjoint

print (R_adjoint - R_forward)/R_forward * 100., "%"
```
